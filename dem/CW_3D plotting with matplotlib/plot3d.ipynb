```python
import numpy as np
xyz = np.array(np.random.random((100, 3)))
```

```python
x = xyz[:, 0]
y = xyz[:, 1]
z = xyz[:, 2]*100
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
pnt3d=ax.scatter(x,y,z,c=z)
cbar=plt.colorbar(pnt3d)
cbar.set_label("Values (units)")
plt.show()
```

```python
# from scipy import misc
import scipy.misc
import imageio
```

```python
dem = imageio.imread('dem.tif')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dem.shape
```

```python
import numpy as np
```

```python
ny, nx = dem.shape
```

```python
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
```

```python
xv, yv = np.meshgrid(x, y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
dem3d = ax.plot_surface(xv, yv, dem)
plt.show()
```

```python
dem3d = ax.plot_surface(xv, yv, dem, cmap='afmhot')
ax.set_title('DEM')
ax.set_zlabel('Elevation (m)')
plt.show()
```

```python
dem3d = ax.plot_surface(xv, yv, dem, cmap='afmhot', linewidth=0)
```

```python
dem3d = ax.plot_surface(xv, yv, dem, cmap='afmhot', linewidth=0, alpha=0.2)
```

```python
dem_100 = dem/100
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
dem3d = ax.plot_surface(xv, yv, dem_100, cmap='afmhot', linewidth=0)
ax.set_title('DEM')
ax.set_zlabel('Elevation (x$10^{-2}$ m)')
plt.show()
```
